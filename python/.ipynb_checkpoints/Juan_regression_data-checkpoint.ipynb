### AUXILIAR FUNCTIONS

```python
#knowing the year build, we can calculate the age of the house, wich i will be more useful for creating themodel.
def house_age(x):
    return int(2020 - int(x))

#datetime wont be useful for our model, thats why why convert it into years and months
def get_years(x):
    x = x.strftime("%Y")
    return int(x)

def get_months(x):
    x = x.strftime("%m")
    return int(x)

#NOT USED ON THE FINAL MODEL, Created to transform the date into seasons
def season_sold(x):
    if x == 3 or x == 4 or x == 5:
        return 1
    elif x == 6 or x == 7 or x == 8:
        return 2
    elif x == 9 or x == 10 or x == 11:
        return 3
    elif x == 12 or x == 1 or x== 2:
        return 0

#SIGN parameters, so wee dont have a lot of low value columns
def is_there_anything(x):
    if x > 0:
        return int(1)
    else:
        return int(0)

#Get the price of one sqft. This might help for predicting the model. WE CANT USE THE ACTUAL PRICE
def sqft_overall(sqft, price):
    return float(sqft / price)

#Sum columns, for example sqft above and sqft basement.
def sum_col(x,y,p):
    z = []
    for i in x:
        z.append((x[i] + y[i])/ p[i])
    return z ;

#Group long and lat by hexagons to afterward we can assign a mean prcie house vlaue by hexagon
def hex_3_point(lat, long):
    value = h3.geo_to_h3(lat=lat, lng = long, resolution = 9)
    return value

#Create a dictionary with the mean price of each hexagon related to the hexagon.
def create_dict(x,y):
    z = x
    z['price'] = y
    newData = z['price'].groupby(z['hex_3']).mean()
    avg_dict_ = newData.to_dict()
    return avg_dict_

#Apply that dictionary to a new dataset.
def apply_dict (x, d):
    x['avg_price'] = range(0,len(x.hex_3))

    for i in range(0,len(x['hex_3'])):
        x['avg_price'][i] = avg_dict[x['hex_3'][i]]
    return x
```

### ORDER FUNCTIONS - CLEANING

```python
#APPLY all those auxiliar functions to the data set we want to clean.
def cleaning_data_0 (df):
    
    #CLEAN THE FRAME
    df.columns=list(map(lambda x: x.lower(),df.columns))
    df.columns=list(map(lambda x: x.replace(' ','_'), df.columns))
    
    #print(df.columns)
    #AGE OF THE HOUSE INTO YR BUILT
    df['house_age'] = df.yr_built.apply(house_age)
    #DATE INTO MONTHS
    df['month_sold'] = list(map(get_months, df['date']))
    #MONTHS INTO SEASONS
    #df['season'] = df.month_sold.apply(season_sold)
    #DUMMIFY IT
    #df = pd.get_dummies(df,columns=["season"])
    #GET THE YEARS
    df['year_sold'] = list(map(get_years, df['date']))
    #ROUND BATHROOMS
    df['bathrooms'] = list(map(round, df['bathrooms']))
    #ENCODE THE BASEMENT
    df['sqft_basement'] = df['sqft_basement'].apply(is_there_anything)
    #ENCODE IF IT IS RENOVATED
    df['yr_renovated'] = df['yr_renovated'].apply(is_there_anything)
    #MAKE FLOORS INTEGERS
    df['floors'] = df['floors'].apply(lambda x: int(x))
    #CREATE HEXAGON WITH LAT AND LONG
    df['hex_3'] = df.apply(lambda x: hex_3_point(x['lat'], x['long']), axis=1)
    df["hex_3"] = df["hex_3"].astype('category')
    df["hex_3"] = df["hex_3"].cat.codes.astype(int)
    
    return df

#Calculate the mean prices of the houses by hexagon and create new values as: sqft_living_price, sqft_lot_price...
#This is only for the train set, because we can only use the prices in the training set. That why we save it afterwards.
def average_prices_hex_train(df, y):
    
    #add the target
    df['avg_price'] = y
    #group prices by hexagons(without sorting)
    dataf = pd.DataFrame(df['avg_price'].groupby(df['hex_3']).mean(), columns = ['avg_price'])
    #sort hexagons by price
    dataf = dataf.sort_values('avg_price',ascending=False)
    dataf = dataf.reset_index()
    #rename columns hexagon to be prepared to set it to the real df
    #dataf = dataf.rename(columns={'hex_3': 'hex_1'})
    #reassign the 
    #df['hex_3'] = dataf['hex_3']
    #make the dict
    #avg_dict = 
    avg_dict = dict(zip(dataf.index.unique() , dataf['avg_price']))
    ind_dict = dict(zip(df.hex_3.unique() , dataf.index.unique()))
    #print(ind_dict)
    #print(dataf)
    #print(avg_dict)
    #print(df['hex_3'])
    #print(df.avg_price.value_counts())
    
    df['hex_3'] = list(map(lambda x: ind_dict[x],df['hex_3']))
    
    df['avg_price'] = list(map(lambda x: avg_dict[x],df['hex_3']))

    #df["hex_3"] = df.sort_values('avg_price',ascending=False)['hex_3']
    
    #df = apply_dict(df,avg_dict)
    df['sqft_living_price'] = df.apply(lambda x: sqft_overall(x['sqft_living'], x['avg_price']), axis=1)
    df['sqft_lot_price'] = df.apply(lambda x: sqft_overall(x['sqft_lot'], x['avg_price']), axis=1)
    df['sqft_living15_price'] = df.apply(lambda x: sqft_overall(x['sqft_living15'], x['avg_price']), axis=1)
    df['sqft_lot15_price'] = df.apply(lambda x: sqft_overall(x['sqft_lot15'], x['avg_price']), axis=1)
    #df['house_sqft_price'] = sum_col(df['sqft_living'],df['sqft_basement'], df['avg_price'])
    
    
    #df.drop('price', inplace = True)
    return df, avg_dict, ind_dict

#Apply de dictionary we created with the trainign set to the test set. And in the same way compute the sqft_living_price, sqft_lot_price...
def average_prices_hex_test(df, dicti, ind_dict):
   
    df['hex_3'] = list(map(lambda x: ind_dict[x],df['hex_3']))
    df['avg_price'] = list(map(lambda x: dicti[x],df['hex_3']))            
    #df = apply_dict(df)
    #CREATE NEW VALUES WITH THE CREATED ONES
    df['sqft_living_price'] = df.apply(lambda x: sqft_overall(x['sqft_living'], x['avg_price']), axis=1)
    df['sqft_lot_price'] = df.apply(lambda x: sqft_overall(x['sqft_lot'], x['avg_price']), axis=1)
    df['sqft_living15_price'] = df.apply(lambda x: sqft_overall(x['sqft_living15'], x['avg_price']), axis=1)
    df['sqft_lot15_price'] = df.apply(lambda x: sqft_overall(x['sqft_lot15'], x['avg_price']), axis=1)
    #df['house_sqft_price'] = sum_col(df['sqft_living'],df['sqft_basement'], df['avg_price'])

    
    return df

#Preform some cleaning that must be done at the end of the cleaning process... Such as dropping columns, or changing the index.
def final_clean(df, dropping):
    df = df.set_index('id')
    if 'price' in df.columns:
        df.drop('price', axis = 1, inplace = True)
        
    df.drop(dropping, axis = 1, inplace = True)
    return df
```

### ORDER FUNCTIONS - TRANSFORMING AND MODELING FUNCTIONS

```python
#Compute the MAPE for the metrics of the model.
def mean_absolute_percentage_error(y_true, y_pred): 
    perc_errors = np.abs((y_true - y_pred) / y_true) * 100
    mean_perc_error = np.mean(perc_errors)
    return  mean_perc_error, perc_errors

#Scale the data direcly with a scling method. Then return the scaler so we can use it afterwards with the test set.
def scale_data(x,scaler):
    ##FIT THE SCALER WITH THE DATA
    optimus_prime = scaler.fit(x)
    ##TRANSFORM IT  WITH THE FIT TRANSFORMER = OPTIMUSPRIME
    data_transf = optimus_prime.transform(x)
    ##CONVERT INTO A DAT FRAME
    data_scaled = pd.DataFrame(data_transf, columns = x.columns)
    print(scaler)
    return data_scaled, optimus_prime

#Test the metric of the model with the testing results
def test_model_metrics(fitmodel, test_x,test_y):
    #Make predictions with the model and the new data set
    my_predictions = fitmodel.predict(test_x)
    #Reverse the log transform.
    my_predictions = np.exp(my_predictions)
    test_y = np.exp(test_y)
    #check results
    MAE =  mean_absolute_error(test_y,my_predictions)
    MSE = mean_squared_error(test_y,my_predictions)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(test_y,my_predictions)
    MPE, ind_perc_errors = mean_absolute_percentage_error(np.asarray(test_y), my_predictions)
    
    
    MAE = "%6.2f" % MAE
    MSE = "%6.2f" % MSE
    RMSE = "%6.2f" % RMSE
    R2 = "%4.2f" % R2
    MPE = "%4.2f" % round(MPE, 4)
    
    
    return my_predictions, MAE, MSE, RMSE, R2, MPE

#Save the model info and metrics into a database
def save_model(df_, model_, scaler_,mae_,mse_,mrse_,r2_,mpe_, desc_):
    #create new df with new data
    datatoadd = pd.DataFrame(np.array([[(model_), (scaler_),(mae_),(mse_),(mrse_),(r2_),(mpe_),(desc_)]]), columns = df_.columns)
    #append both dataframes
    new_df = pd.concat([df_, datatoadd])
    return new_df

def save_model_intodb(model,scaler,mae, mse, mrse, r2, mpe, desr):
    new_df = save_model(trained_saved_models,str(model),str(scaler),mae, mse, mrse, r2, mpe, desr)
    return new_df
```

### EXECUTING FUNCTIONS - TRAIN AND TEST 

```python
#DEFINITION TO TRAIN, TEST, AND THEN SAVE RESULTS OF THE MODEL
def train_model(x,y, model,scaler):
    #CLEAN DATA
    x = cleaning_data_0(x)
    #CREATE DICTIONARY
    x, avg_dict, hex_dict = average_prices_hex_train(x,y)
    #apply the dictionary
    #x = average_prices_hex_test(x,avg_dict)
    #final cleaning
    x = final_clean(x, list_to_drop)
    #SCALE DATA AND SAVE THE SCALER
    scaled_data, optimusprime = scale_data(x, scaler)
    #FIT THE SCALED DATA INTO THE MODEL
    fit_model = model.fit(scaled_data, y)
    return x,fit_model, optimusprime,avg_dict, hex_dict

#DEFINITION TO TEST THE MODELS WE USED
def test_model(xtest, ytest, testing_model, trained_scaler, diction, hex_dict):
    #CLEAN DATA
    xtest = cleaning_data_0(xtest)
    #APPLY DICTIONARY
    xtest = average_prices_hex_test(xtest,diction,hex_dict )
    #final cleaning
    xtest = final_clean(xtest,list_to_drop)
    #Scale the test data with the trained scaler
    xtest = trained_scaler.transform(xtest)
    #TEST THE MODEL RESULTS
    _my_predictions, mae, mse, mrse, r2, mpe = test_model_metrics(testing_model,xtest,ytest)
    #RESPRESENT THE DATA
    print ("MAE: " + mae + " MSE: " + mse +" MRSE: " + mrse +" R2: "+ r2 +" MPE: "+ mpe)
    
    fig, ax = plt.subplots()

    ax.scatter(np.exp(ytest), _my_predictions)
    ax.plot([np.exp(ytest).min(), np.exp(ytest).max()], [np.exp(ytest).min(),np.exp(ytest).max()], 'k--', lw=4)
    ax.set_xlabel('Measured')
    ax.set_ylabel('Predicted')
    
    sns.scatterplot(y = _my_predictions, x = np.exp(ytest))
    return xtest,mae, mse, mrse, r2, mpe
```

```python
def corr_matrix(df):  
    corr_matrix=df.corr(method='pearson')  # default
    mask = np.zeros_like(corr_matrix)
    mask[np.triu_indices_from(mask)] = True
    fig, ax = plt.subplots(figsize=(10, 8))
    ax = sns.heatmap(corr_matrix,mask = mask, annot=True)
    plt.show()
```

```python
def columns_vif(df, target, drop=[]):
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    temporary_vif_frame = df.drop(drop, axis=1)
    dropped_vif_frame = temporary_vif_frame.drop(target, axis=1)
    cols = dropped_vif_frame.columns
    temporary_cleaned_data = pd.DataFrame.to_numpy(dropped_vif_frame)
    vif = [variance_inflation_factor(temporary_cleaned_data, i) for i in range(temporary_cleaned_data.shape[1])]
    for i in range(len(cols)):
        print(str(cols[i]) + "--------" + str(vif[i]))
    return temporary_vif_frame
```
