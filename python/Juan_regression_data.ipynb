```python
def save_model(df_, model_, scaler_,mae_,mse_,mrse_,r2_,mpe_):
    #create new df with new data
    datatoadd = pd.DataFrame(np.array([[(model_), (scaler_),(mae_),(mse_),(mrse_),(r2_),(mpe_)]]), columns = df_.columns)
    #append both dataframes
    new_df = pd.concat([df_, datatoadd])
    return new_df
```
